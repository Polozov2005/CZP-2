**itertools**

itertools - это встроенная библиотека для работы с итераторами и создания сложных итераторов. Она оптимизирована для эффективной работы с большими наборами данных.

Итераторы:

```python
# product() - декартово произведение
result = itertools.product([1, 2], ['a', 'b'])
print(list(result))  # [(1, 'a'), (1, 'b'), (2, 'a'), (2, 'b')]

# permutations() - перестановки
perms = itertools.permutations([1, 2, 3], 2)
print(list(perms))  # [(1, 2), (1, 3), (2, 1), (2, 3), (3, 1), (3, 2)]

# combinations() - комбинации (без повторений)
combs = itertools.combinations([1, 2, 3, 4], 2)
print(list(combs))  # [(1, 2), (1, 3), (1, 4), (2, 3), (2, 4), (3, 4)]
```
